# New Section

```python
# Read in CSV file
fake_jobs_df = pd.read_csv("../content/sample_data/fake_job_postings.csv")
fake_jobs_df
```

```python
# Check for entirely duplicated rows
fake_jobs_df = fake_jobs_df.drop_duplicates()
fake_jobs_df
```

```python
# Check the duplicate descriptions to get an idea
fake_jobs_df.sort_values("description").head(60)
```

```python
# Create column to assess whether or not salary range is provided
fake_jobs_df['salary_provided'] = pd.notna(fake_jobs_df['salary_range'])
fake_jobs_df['benefits_provided'] = pd.notna(fake_jobs_df['benefits'])

fake_jobs_df
```

```python
# Check datatypes
fake_jobs_df.dtypes
```

```python
# Convert Boolean columns to integer
fake_jobs_df[['salary_provided', 'benefits_provided']] = (fake_jobs_df[['salary_provided', 'benefits_provided']]).astype(int)
fake_jobs_df
```

```python
# Separate Location into Country, State, City
fake_jobs_df['country'], fake_jobs_df['state/province'], fake_jobs_df['city'] = fake_jobs_df['location'].str.split(',', 2).str
fake_jobs_df
```

```python
# Reorganize Columns
fake_jobs_df = fake_jobs_df[['job_id', 'city', 'state/province', 'country', 'title', 'department', 'industry', 
                             'function', 'salary_range', 'salary_provided', 'company_profile', 'description', 
                             'requirements', 'benefits', 'benefits_provided', 'telecommuting', 'has_company_logo', 
                             'has_questions', 'employment_type', 'required_experience', 'required_education', 
                             'fraudulent']]
fake_jobs_df
```

```python
# Save Cleaned/Organized df to csv
fake_jobs_df.to_csv("fake_jobs_clean.csv", index=False)
```
